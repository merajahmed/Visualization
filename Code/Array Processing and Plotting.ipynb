```python
%pylab inline

from numpy import *
from matplotlib import *
from matplotlib.pyplot import *

#1. Create a numpy array with 100 elements from 1 to 100 in order :  use box plot to draw the data you generate
boxarray = array(arange(1,101))
figure()
boxplot(boxarray)
show()
```

```python
#2. Create a nump array with 8000 random numbers.  display a histogram of the data using 20 bins 
randarray = random.uniform(0,1,8000);
figure()
hist(randarray,20)
show()
```

```python
#3. Reshape it into a 20 x 20 x 20 array, and with the reshaped numpy array, compute the average values for every x,y,and z slice
randarray.shape = (20,20,20)
print 'means for x slices:-'
for i in range(0,20):
    print 'mean for x slice ',i+1, ' : ',mean(randarray[i,:,:])
print 'means for y slices:-'
for i in range(0,20):
    print 'mean for y slice ',i+1, ' : ',mean(randarray[:,i,:])
print 'means for z slices:-'
for i in range(0,20):
    print 'mean for z slice ',i+1, ' : ',mean(randarray[:,:,i])
```

```python
#4. Average all the x slices together and display a single 2D image of the averaged x slice. Do the same for y and z slices  
xslice_mean = mean(randarray, axis=0)
yslice_mean = mean(randarray, axis=1)
zslice_mean = mean(randarray, axis=2)
imshow(xslice_mean)
print 'Averaged x slice image'
show()
imshow(yslice_mean)
print 'Averaged y slice image'
show()
print 'Averaged z slice image'
imshow(zslice_mean)
show()
```

```python
#Write a python program, generate 100 random number uniformly distributed between 1 and 100 , write the numbers out to a binary file : use a line graph to draw the 100 numbers
from numpy import *
from matplotlib import *
from matplotlib.pyplot import *

a = random.uniform(1, 100, 100)
save("randomNum.npy",a)
figure()
plot(a)
grid()
show()
```

```python
#Write a python program to read the binary file back, divide the range between 1 and 100 into 7 intervals, and calculate the frequency for each interval: display a histogram of your result
a = load("randomNum.npy")
bins = [0] * 7
for i in a:
	index = int((i - 1) / 100 * 7)
	bins[index] = bins[index] + 1
hist(a)
show()
```

```python
#Write a python function that takes a list of random numbers as input, compute the mean and the standard distribution: draw a line graph for your data 
def getStats(alist):
	print ("mean: " + str(mean(alist)))
	print ("std: " + str(std(alist)))
	figure()
	plot(alist)
	grid()
	show()
```
